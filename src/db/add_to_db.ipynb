```python
lst:str=[]

#* Rebource -> https://kuran.diyanet.gov.tr/Yayinlar
with open("data/doc/Kuran.txt","r") as f:
    lst=f.readlines()
    
#? filter data
lst.remove("﷽\n")



with open("data/doc/cikti.txt","w") as f:
    f.writelines(lst)
```

## add to database

```python
sure_id=0
sayfa=0
for i in lst:
    if "﷍	﴿" in i: #? sure isimleri
        sure_id+=1
        db_m.add_ayat(-1,i,sure_id,0,sayfa+1)
        #print(-1,i,sure_id,0)
        ayet_id=1
    elif "﴾" in i:
        sayfa+=1
        ayetler=i.split("﴾")
        ayetler.pop()
        for a in ayetler:
            db_m.add_ayat(ayet_id,a+"﴾",sure_id,0,sayfa)
            #print(ayet_id,a+"﴾",sure_id,0)
            ayet_id+=1
db_m.end_process()
```
